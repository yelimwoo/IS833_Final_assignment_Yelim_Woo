**Data source:**

Our data is from the Stack Overflow Annual Developer Survey based on a survey of 13,023 software developers from 180 countries around the world. The annual developer survey examines all aspects of the developer experience from learning to code to their favorite technologies to version control and the workplace experience of professional developers. The survey was fielded from May 11, 2022, to June 1, 2022. Respondents were recruited primarily through channels owned by Stack Overflow. The top sources of respondents were onsite messaging, blog posts, email lists, banner ads, and social media posts. Since respondents were recruited in this way, highly engaged users on Stack Overflow were more likely to notice the prompts to take the survey over the duration of the collection promotion. As an incentive, respondents who finished the survey could opt into a “Census” badge if they completed the survey.

# **Challenges**

Following challenges were faced by our team:


1. The original dataset had a large number of columns, which was not suitable for the purposes of data analysis. Hence, we had to omit several columns to make the analysis feasible and less complex. We also used a subset of the original dataset having a salary range of 50,000 to 200,000 USD.

2. Several fields in the original dataset had missing values, which needed to be handled appropriately to make the final analysis make sense.

3. Even after omitting several columns, it was essential to combine some of the related variables into a new variable, for the sake of simplicity of analysis.

4. Some parameters , such as “MiscTechHaveWorkedWith”- other frameworks and libraries  the developer has worked with , had a large number of null values which could have an effect on the insights made. 

5. Multiple choice answers such as languages, databases, platforms that the developers use are very complex and complicated. Each of these fields had multiple choices, resulting in thousands of combinations, which was extremely difficult to analyze.

  For example, separating the values into each language was very difficult and it was not possible to use one hot encoding. We tried to separate C and C++, and made a new column named C with an input value of 1. Since C is included in C++, it took a lot of time to encode the language. So, we only focused on a few popular languages and programs to analyze the differences in salary.
